##### Merging the datasets

```python
# imports

import pandas as pd
import numpy as np
```

--------------------

```python
D1 = pd.read_csv('DATA/genome-scores.csv', index_col=False)
D1.head()
```

```python
D2 = pd.read_csv('DATA/genome-tags.csv', index_col=False)
D2.head()
```

```python
D3 = pd.read_csv('DATA/links.csv', index_col=False)
D3.head()
```

```python
D4 = pd.read_csv('DATA/movies.csv', index_col=False)
D4.head()
```

```python
D5 = pd.read_csv('DATA/ratings.csv', index_col=False)
D5.head()
```

```python
D6 = pd.read_csv('DATA/tags.csv', index_col=False)
D6.head()
```

```python
merge1 = pd.merge(D1, D2, on='tagId', how='outer')
merge1.head()
```

```python
merge2 = pd.merge(merge1, D3, on='movieId', how='outer')
merge2.head()
```

```python
merge2.sample(n=5, random_state=1)
```

```python
merge2[merge2['tagId'] == 1.0]
```

```python
merge3 = pd.merge(merge2, D4, on='movieId', how='outer')
merge3.head()
```

```python
merge4 = pd.merge(merge3, D5, on='movieId', how='outer')
merge4.head()
```
